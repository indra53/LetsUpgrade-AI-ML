# Predict House Pricing using Linear Regression

```python
import pandas as pd

import numpy as np

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
data=pd.read_excel('Linear Regression.xlsx',axis=0)
```

```python
data.head()
```

```python
data.corr() # checking corelation with each data
```

```python
sns.pairplot(data)

# price as corelation with sqft and other variables too have corelation with each other that is good for model
```

# Split dataset in X and Y

```python
x=data[['sqft_living', 'bedrooms', 'bathrooms', 'floors']]

y=data.price
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=100)
```

# Training the Model

```python
from sklearn.linear_model import LinearRegression
```

```python
Lin_model=LinearRegression()
```

```python
train=Lin_model.fit(x_train,y_train)
```

# Checking Coefficient

```python
coef=pd.DataFrame(train.coef_,x.columns,columns=['Coefficient'])
```


# coefficent mean chnaging one unit in x(independent varibale) associate with y(Dependent variable)

hence:  we can say one unit chnage in sqft_living value will chnage price by 304.930204
        we can say one unit chnage in bedrooms value will chnage price by -54729.625300(it has negative correlation
     
        we can say one unit chnage in bathrooms value will chnage price by 9916.921300
        
        we can say one unit chnage in floors value will chnage price by 2417.837868

```python
coef
```

# Predict house Price

```python
prediction=Lin_model.predict(x_test)
```

```python
prediction
```
